## My Quantum Computing Journey with PennyLane

## Introduction to Quantum Computing

### Understanding Qubits
I learned the fundamental concept of qubits, the basic unit of quantum information. Key takeaways included:
- **Mathematical Representation:** Qubits can be represented using Dirac notation (bra-ket) and as vectors in a 2-dimensional complex vector space.
- **Superposition:** A qubit can exist in multiple states simultaneously, described by a linear combination of basis states |0⟩ and |1⟩.
- **Normalization and Measurement:** The probabilities of measurement outcomes are derived from the amplitudes of the qubit's state, ensuring that the total probability sums to 1.

### Codercises Completed:
- **Normalization of quantum states:** Practiced normalizing various quantum states.
- **Inner product and orthonormal bases:** Calculated inner products and explored orthonormal bases.
